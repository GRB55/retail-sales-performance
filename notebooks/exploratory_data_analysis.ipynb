```python
# Data manipulation
import pandas as pd

pd.set_option("display.max_columns", None)
```

```python
# First excel sheet with 2009-2010 data
df_2009_2010 = pd.read_excel(r"C:\Users\Gonzalo\Desktop\Archivos python\Retail-Sales-Performance\data\raw\online_retail_II.xlsx", sheet_name="Year 2009-2010")
# Second excel sheet with 2010-2011 data
df_2010_2011 = pd.read_excel(r"C:\Users\Gonzalo\Desktop\Archivos python\Retail-Sales-Performance\data\raw\online_retail_II.xlsx", sheet_name="Year 2010-2011")

# Show the datasets
display(df_2009_2010)
display(df_2010_2011)
```

```python
# Values that should not exist
df[df["Quantity"] <= 0]
```

```python
# Values that should not exist
df[df["Price"] <= 0]
```

```python
# Cancellations
df[df["Invoice"].str.startswith("C", na=False)]
```

```python
# Cancellations and returns
df[(df["Invoice"].str.startswith("C", na=False)) &
   (df["Quantity"] <= 0)]
```

```python
# No product nor price nor customer ID
df[(df["Description"].isna()) &
   (df["Price"] <= 0) &
   (df["Customer ID"].isna())]
```

```python
# Stock code with no letters
df[df["StockCode"].astype(str).str.match(r"^\d+$", na=False)]
```

```python
# Stock code including letters at the end
df[~df["StockCode"].astype(str).str.match(r"^\d+$")]
```

```python
# Stock code for discounts, shipping, fees, etc.
df[df["StockCode"].astype(str).str.match(r"[a-z-A-Z]")]["StockCode"].value_counts().head(10)
```
